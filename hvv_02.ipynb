```python
leftjoin_df = stop_times_df.merge(stops_df, how='left', on='stop_id')
leftjoin_df
```

```python
clean_df = leftjoin_df.drop(["pickup_type", "drop_off_type", "stop_headsign", "stop_code", "stop_desc", "location_type", "parent_station", "wheelchair_boarding", "platform_code", "arrival_time", "departure_time"], axis = 1)
```

```python
clean_df = clean_df[["trip_id", "stop_sequence", "stop_id", "stop_name", "stop_lat", "stop_lon" ]]
# clean_df
```

```python
# make a tuple out of a stop's lat and long 
# clean_df['lat_long'] = list(zip(clean_df.stop_lat, clean_df.stop_lon))
# clean_df
```

```python
# Create new columns with the shifted values of the latitude and longitude columns
clean_df['lat_shift'] = clean_df['stop_lat'].shift(1)
clean_df['lon_shift'] = clean_df['stop_lon'].shift(1)
clean_df
```

```python
import numpy as np

def haversine(lon1, lat1, lon2, lat2):                                                                                  
    ''' Returns the distance between the points in km. Vectorized and will work with arrays and return an array of       
    distances, but will also work with scalars and return a scalar.'''                                                  
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])                                                  
    a = np.sin((lat2 - lat1) / 2.0)**2 + (np.cos(lat1) * np.cos(lat2) * np.sin((lon2 - lon1) / 2.0)**2)                 
    distance = 6371 * 2 * np.arcsin(np.sqrt(a))                                                                         
    return distance

clean_df['distance'] = haversine(clean_df.stop_lon, clean_df.stop_lat, clean_df.lon_shift, clean_df.lat_shift)
clean_df
```

```python
# calculate the distance between two points using geopy to cross check values from above df
from geopy.distance import distance

point1 = (53.876897,	9.889825)
point2 = (53.904000,	9.890462)

print(distance(point1, point2).km)
```

```python
clean_df.loc[clean_df['stop_sequence']==0, ['distance']] = 0
# clean_df.head(50)
```

```python
grouped_sum = clean_df.groupby(["trip_id"], sort = False).agg({'distance':'sum'})
grouped_sum
```

```python
trips_df = pd.read_csv("trips.txt")
grouped_sum_lj = trips_df.merge(grouped_sum, how='left', on='trip_id')
```
